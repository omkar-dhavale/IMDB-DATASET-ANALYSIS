#Q1)name all the directors with the number of movies and revenue in order of revenue

```python
cur.execute("select name, count(title),sum(revenue) from directors join movies on directors.id = movies.director_id group by director_id order by sum(revenue) desc  ")
omkar=cur.fetchall()
mix_df=pd.DataFrame(omkar,columns=["name","Total_movies","Revenue"])
mix_df
```

#Q2).name all the directors with the number of movies and revenue in order of no. of movies
